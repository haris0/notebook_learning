## Soal 1
Dari Dataset iris, buat sebuah kolom baru yang berisikan sebuah penjelasan mengenai Panjang sepal dan variasi pada setiap data

```python
import pandas as pd
file_path = "iris_dataset.csv"
dataf = pd.read_csv(file_path)
ket = []
for i, row in dataf.iterrows():
    ket.append('bunga ke {} merupakan variasi {} dengan sepal length {}'.format(i, row[4], row[0]))
dataf['keterangan'] = pd.Series(ket, index=dataf.index)
dataf
```

## Soal 2
Buat 3 histogram berbeda pada satu plot yang berisi Panjang sepal pada masing-masing variasi dilengkapi dengan garis distribusi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Setosa = dataf.loc[dataf["variety"]=="Setosa", "sepal.length"]
Versicolor = dataf.loc[dataf["variety"]=="Versicolor", "sepal.length"]
Virginica = dataf.loc[dataf["variety"]=="Virginica", "sepal.length"]

fig, axes = plt.subplots(1, 3, figsize=(9,6), sharey=True)
sns.distplot(Setosa, color='red', axlabel='Satosa', hist_kws={"alpha":0.5}, ax=axes[0])
sns.distplot(Versicolor, color='green', axlabel='Versicolor', hist_kws={"alpha":0.5}, ax=axes[1])
sns.distplot(Virginica, color='blue', axlabel='Virginica', hist_kws={"alpha":0.5}, ax=axes[2])

plt.suptitle('Sepal Lenght')
plt.show()
```

## Soal 3
Dari dataset iris, pilihlah dua fitur (kolom) yang paling penting dan tampilkan kedua fitur tersebut

```python
from sklearn.feature_selection import SelectKBest

bestTwo = SelectKBest(k=2)
y = dataf['variety']
x = dataf.drop(['variety','keterangan'], axis=1)

fit = bestTwo.fit(x, y)
print(fit.scores_)
features = fit.transform(x)
print(features)
```

## Soal 4
Dari data â€œiris_dataset_missing.csvâ€, ubahlah :
##### Semua nilai 0 yang terdapat pada kolom â€œsepal.lengthâ€ menjadi nilai rata-rata â€œsepal.lengthâ€
##### Semua nilai 0 yang terdapat pada kolom â€œsepal.widthâ€ menjadi nilai modus â€œsepal.widthâ€
##### Semua nilai 0 yang terdapat pada kolom â€œpetal.lengthâ€ menjadi nilai median â€œpetal.lengthâ€
##### Semua nilai 0 yang terdapat pada â€œpetal.widthâ€ menjadi 5.

```python
from sklearn.impute import SimpleImputer

file_path = "iris_dataset_missing.csv"
datafM = pd.read_csv(file_path)

impute_sl = SimpleImputer(missing_values=0, strategy='mean')
impute_sl.fit(datafM[['sepal.length']])
impute_sl.transform(datafM[['sepal.length']])
datafM['sepal.length'] = impute_sl.transform(datafM[['sepal.length']])

impute_sw = SimpleImputer(missing_values=0, strategy='most_frequent')
impute_sw.fit(datafM[['sepal.width']])
impute_sw.transform(datafM[['sepal.width']])
datafM['sepal.width'] = impute_sw.transform(datafM[['sepal.width']])

impute_pl = SimpleImputer(missing_values=0, strategy='median')
impute_pl.fit(datafM[['petal.length']])
impute_pl.transform(datafM[['petal.length']])
datafM['petal.length'] = impute_pl.transform(datafM[['petal.length']])

impute_pw = SimpleImputer(missing_values=0, strategy='constant', fill_value=5)
impute_pw.fit(datafM[['petal.width']])
impute_pw.transform(datafM[['petal.width']])
datafM['petal.width'] = impute_pw.transform(datafM[['petal.width']])

datafM
```
