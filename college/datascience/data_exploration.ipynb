```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('http://www.sharecsv.com/dl/2c5b0314d0f61f3e4615ee9289c973e1/Tugas4.csv')
df
```

## Soal 1

```
print("Income :")
print('Mean \t\t: ', df['Income'].mean())
print('Standar deviasi\t: ', df['Income'].std())
print('Median \t\t: ', df['Income'].median())
mode0, mode1 = df['Income'].mode()
print('Mode 1\t\t: ', mode0)
print('Mode 2\t\t: ', mode1)
```

```
print("\nAge :")
print('Mean \t\t: ', df['Age'].mean())
print('Standar deviasi\t: ', df['Age'].std())
print('Median \t\t: ', df['Age'].median())
mode0 = df['Age'].mode()
print('Mode 1\t\t: ', mode0)
```
